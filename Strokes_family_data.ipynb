# Import data 

```r
data <- read.csv("healthcare-dataset-stroke-data - healthcare-dataset-stroke-data1.csv")
head(data)
```

# Import libraries

```r
# Libraries to acess if data is parametic
library("dplyr")  #Returns a value
library("ggpubr") #Data visualization
```

# Cleaning data

```r
#Bmi came as a character data, so I've changed it
sum(is.na(data$bmi))
data$bmi <- as.numeric(data$bmi)
```

# Basic statistics

```r
summary(data)
```

**Media close to median:** age,bmi
**Media far from median:** avg_glucose_level

```r
#Visual normality test on age
ggdensity(data$age,
         main = "Density plot of age",
         xlab = "age")

ggqqplot(data$age,
        main = "Q-Q plot of age",
        xlab = "age")
```

```r
#Visual normality test on Average glucose level
ggdensity(data$avg_glucose_level,
         main = "Density plot of average glucose level",
         xlab = "Average glucose level")

ggqqplot(data$avg_glucose_level,
        main = "Q-Q plot of average glucose level",
        xlab = "Average glucose level")
```

```r
#Visual normality test on BMI
ggdensity(data$bmi,
         main = "Density plot of BMI",
         xlab = "BMI")

ggqqplot(data$bmi,
        main = "Q-Q plot of BMI",
        xlab = "BMI")
```
